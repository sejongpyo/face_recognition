```python
test_img = cv2.imread('images/iron.jpg')
test_img_gray = cv2.cvtColor(test_img, cv2.COLOR_BGR2GRAY)
plt.imshow(test_img_gray, cmap='gray')
```

```python
def convert_RGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
haar_cascade_face = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
faces_rects = haar_cascade_face.detectMultiScale(test_img_gray, scaleFactor = 1.3, minNeighbors = 5)

print(len(faces_rects))
```

```python
for (x, y, w, h) in faces_rects:
    cv2.rectangle(test_img, (x, y), (x+w, y+h), (0, 255, 0), 2)

plt.imshow(convert_RGB(test_img))
```

```python
def detect_faces(cascade, test_image, scaleFactor = 1.1):
    image_copy = test_image.copy()
    gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)
    faces_rect = cascade.detectMultiScale(gray_image, scaleFactor = scaleFactor, minNeighbors = 5)
    for (x, y, w, h) in faces_rect:
        cv2.rectangle(image_copy, (x, y), (x+w, y+h), (0, 255, 0), 15)
        
    return image_copy
```

```python
test_image2 = cv2.imread('images/thor.jpg')

test_image2_gray = cv2.cvtColor(test_image2, cv2.COLOR_BGR2GRAY)

plt.imshow(test_image2_gray, cmap = 'gray')
```

```python
faces = detect_faces(haar_cascade_face, test_image2)
plt.imshow(conver_RGB(faces))
```
